```python
# ... pip install   ...

import pandas as pd
import matplotlib.pyplot as plt

# 1.  
try:
    df = pd.read_csv('Big241227.csv', encoding='utf-8')  # encoding  
except FileNotFoundError:
    print("    .   .")
    exit()

# 2.  
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
monthly_data = df.resample('ME').sum()  # 'M'  'ME' 

# 3.  
plt.figure(figsize=(10, 6))
plt.plot(monthly_data.index, monthly_data['value'], marker='o')
plt.title('Monthly Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.xticks(rotation=45)  # x  
plt.tight_layout()  #  
plt.show()
```
